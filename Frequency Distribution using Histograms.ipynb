```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
# Initialize tokenizer
tokenizer = TreebankWordTokenizer()
# Sample text
text = "Natural Language Processing (NLP) with Python is fun and educational. Python simplifies NLP tasks."
# Tokenize text
tokens = tokenizer.tokenize(text)
# Print tokens
print("Tokens:", tokens)
# Frequency Distribution
fdist = FreqDist(tokens)
# Print frequencies
print("\nWord Frequency Distribution:")
for word, freq in fdist.items():
    print(f"{word}: {freq}")
# Plot frequency distribution
fdist.plot(title='Simple Frequency Distribution Plot')
plt.show()
```

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Download stopwords (if not already downloaded)
nltk.download('stopwords')

# Initialize tokenizer and stopwords list
tokenizer = TreebankWordTokenizer()
stop_words = set(stopwords.words('english'))

# Sample text (fixed line break)
text = "Natural language processing with Python is interesting and useful. Python makes NLP tasks easier and more efficient."

# Tokenize the text
tokens = tokenizer.tokenize(text)

# Remove stopwords and punctuation
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]

# Print filtered tokens
print("Filtered Tokens (without stopwords):", filtered_tokens)

# Frequency Distribution
fdist = FreqDist(filtered_tokens)

# Print frequency distribution
print("\nWord Frequency Distribution (without stopwords):")
for word, freq in fdist.items():
    print(f"{word}: {freq}")

# Plot frequency distribution
fdist.plot(title='Word Frequency Distribution (Without Stopwords)')
plt.show()
```
